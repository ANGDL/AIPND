## 1. 什么是向量？

向量用一组数表示: [1, 2, 5, 5, 6]。   

一组数有什么用？ 可以用来表示方向。


在课程中（不是几何学）起始点是不重要的，默认是从(0,0)点出发。  空间中的任意一个点，都可以表示成从原点指向这个点的有向线段。



数字的顺序不同，所表示的向量也不相同。

按场景定义向量的维度的意义。


```python
[120, 3, 2, 666, 2]
```

**总结：向量可以理解为一组有序的数字。**

## 2. 向量 vs 标量

1. 标量：只有大小，没有方向，一般用一个数字表示: $a$。
2. 向量：既有大小，又有方向，向量的符号画箭头：$\vec v$

## 3. 行向量 vs 列向量
教材中一般都指列向量。   
数学表示：   
行向量：$[1, 2]$   
列向量：$$\left[
 \begin{matrix}
   1 \\
   2 \\
   3
  \end{matrix}
  \right]
$$

## 4. 向量的加法
$$\begin{bmatrix} v_1 \\ v_2 \\ .\\ . \\ v_3\end{bmatrix} + \begin{bmatrix} w_1 \\ w_2 \\ .\\ . \\ w_3\end{bmatrix} = \begin{bmatrix} v_1+w_1 \\ v_2+w_2 \\ .\\ . \\ v_3+w_3\end{bmatrix}$$


## 5. 向量与数量相乘
$$ k \times \begin{bmatrix} v_1 \\ v_2 \\ .\\ . \\ v_3\end{bmatrix} = \begin{bmatrix} k \times v_1 \\ k \times v_2 \\ .\\ . \\ k \times v_3\end{bmatrix}$$



## 6. 零向量
对于任意一个向量 $\vec u$，都存在一个向量 $O$, 满足：$\vec u + O = \vec u$, $O$ 为零向量。  
PS：零向量没有箭头。

## 向量的长度：向量的模

 
$$||\vec u|| = \sqrt{3^2 + 4^2}$$

对于n维向量：$\vec u = (u_1, u_2, ..., u_n)^T$, $||\vec u|| = \sqrt{u_1^2 + u_2^2 + ... + u_n^2}$

## 7. 单位向量
单位向量只表示方向

对于向量$\vec u = (u_1, u_2, ..., u_n)^T$，    
其单位向量   
$$\hat u = (\frac{u_1}{||\vec u||}, \frac{u_2}{||\vec u||}, ..., \frac{u_n}{||\vec u||})^T$$

$$||\hat u|| = 1$$



## 8. 标准单位向量


二维空间中，有两个标准单位向量： $\vec e_1 = (1, 0)$，$\vec e_2 = (0, 1)$   
三位空间中，有三个标准单位向量：$\vec e_1 = (1, 0, 0)$，$\vec e_2 = (0, 1, 0)$, $\vec e_3 = (0, 0, 1)$   
n位空间中，有n个标准单位向量：$\vec e_1 = (1, 0, ... , 0)$，$\vec e_2 = (0, 1, ... , 0)$, ... , $\vec e_n = (0, 0, ... , 1)$

## 9. 向量的点乘（内积）
参考资料：https://zh.wikipedia.org/wiki/%E7%82%B9%E7%A7%AF   

**两个向量“相乘”，结果是一个数（标量）！**


**不同方向的向量不能直接做乘法，所以先把一个向量投影到一个方向上来，再做长度的乘积。   
点乘反映着两个向量的“相似度”，两个向量越“相似”，它们的点乘越大。**
