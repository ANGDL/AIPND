## 1. --


    
$$\hat y=sigmoid(\vec{w}^T\centerdot\vec{z})\qquad$$   
   
$$sigmoid(z)=\frac{1}{1+e^{-z}}$$
$y=sigmoid(z)$
sigmoid$y'=y(1-y)$ 
  

$$\hat y=\frac{1}{1+e^{-\vec{w}^T\centerdot\vec{x}}}$$



```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pylab as plt
import numpy as np
torch.manual_seed(0)

%matplotlib inline
```

```python
X = torch.tensor([
    [1.0, 3.0, 10.0],
    [2.0, 5.0, 9.0],
])

print(X)
```

```python
z = torch.arange(-10,10,0.1)
sigmod_z = torch.sigmoid(z)
plt.plot(z.numpy(), sigmod_z.numpy())
plt.xlabel('z')
plt.ylabel('sigmod_z')
```

```python
model = nn.Sequential(nn.Linear(3, 1), nn.Sigmoid())
model[0].weight.data = torch.tensor([[0., 0., 0.]])
model[0].bias.data = torch.tensor([0.])
model
y_hat = model(X)
y_hat
```

```python
model
```

```python
model[0].state_dict()
```

## 2. 


$$sofmax(z) = \frac{exp(z^{(i)})}{\sum^C_{j=1}exp(z^{(j)})}$$

$$relu(a) = max(0,x)$$

```python
# pytorchsoftmax
z = torch.arange(-10,10,0.1)
z[20] = 100
softmax_z = torch.softmax(z, dim=0)
plt.plot(z.numpy(), softmax_z.numpy())
plt.xlabel('z')
plt.ylabel('softmax_z')
print(torch.sum(softmax_z))
```

```python
relu_z = torch.relu(z)
plt.plot(z.numpy(), relu_z.numpy())
plt.xlabel('z')
plt.ylabel('relu_z')
```

### 3.2 stride

$$M_{new}=\dfrac{M-K}{stride}+1$$

```python
K=2
stride = 3
conv1 = nn.Conv2d(in_channels=1, out_channels=1,kernel_size=K, stride=stride)
conv1.state_dict()['weight'][0][0]=torch.tensor([[1.0,1.0],[1.0,1.0]])
conv1.state_dict()['bias'][0]=0.0
conv1.state_dict()
conv1
```

```python
M=4
image1=torch.ones(1,1,M,M)
```

```python
z1=conv1(image1)
print("z1:",z1)
print("shape:",z1.shape[2:4])
```

### 3.3 padding

$$M'=M+2 \times padding$$
$$M_{new}=M'-K+1$$

```python
conv2 = nn.Conv2d(in_channels=1, out_channels=1,kernel_size=2,stride=3,padding=2)

conv2.state_dict()['weight'][0][0]=torch.tensor([[1.0,1.0],[1.0,1.0]])
conv2.state_dict()['bias'][0]=0.0
conv2.state_dict()
z2=conv2(image1)
print("z2:",z2)
print("z2:",z2.shape[2:4])
```
